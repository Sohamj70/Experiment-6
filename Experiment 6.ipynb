```python
#6. Write a python program to make increement in the date
# Input date from user
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

# Number of days in each month
days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# Check leap year
if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    days[1] = 29   # February has 29 days in leap year

# Increase day by 1
day = day + 1

# Check if day exceeds days in month
if day > days[month - 1]:
    day = 1
    month = month + 1

    # Check if month exceeds December
    if month > 12:
        month = 1
        year = year + 1

# Print next date
print("Next Date:", day, "-", month, "-", year)


```

```python
'''problem statement:
write a python pogram to calculate the gross salary of an employee based
if basic <=10000: HRA = 20%, DA = 80%
if basic <=20000: HRA = 25%, DA = 90%
otherwise HRA = 30%, DA = 95%'''

basic = int(input("Enter the Basic Salary: "))

if basic <= 10000:
  hra = 0.2 * basic
  da = 0.8 * basic
elif basic <= 20000:
  hra = 0.2 * basic
  da = 0.9 * basic
else:
  hra = 0.3 * basic
  da = 0.9 * basic


# Calculate Gross Salary
gross_salary = basic + hra + da
print(f"Gross Salary: ",gross_salary)
```

```python
'''Write  a python program to calculate income tax based on annual income
upto ₹2,50,000 : No Tax
₹2,50,001 - ₹5,00,000 : 5%
₹5,00,001 - ₹10,00,000 : 20%
above ₹10,00,000 : 30%'''

income = int(input("Enter your annual income (₹): "))
tax = 0
if income <= 250000:
  tax = 0
elif income <= 500000:
  tax = (income - 250000) * 0.05
elif income <= 1000000:
  tax = 12500 + (income - 500000) * 0.20
else:
  tax = 112500 + (income - 1000000) * 0.30
print("tax : ", tax)
```
